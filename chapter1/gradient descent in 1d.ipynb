```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

display.set_matplotlib_formats('svg')
```

### 定义函数

```python
def fx(x):
    return 3*x**2 - 3*x +4

def deriv(x):
    return 6*x - 3
```

```python
# 定义x 的取值范围
x =np.linspace(-2,2,2001)


# 从x中随机选择一个作为初始的局部最小值
localmin = np.random.choice(x, 1)


learning_rate = .01
training_epoches = 100

for i in range(training_epoches):
    # 计算当前点的梯度
    grad = deriv(localmin)
    # 当grad为负数时，localmin - 负数 保证了我们往前走，也就是保证梯度 在下降。 当导数grad为正数时，localmin -正数，在图像上，我们可以看到是在后退.
    localmin = localmin - learning_rate * grad

print(localmin)
```

```python

plt.plot(x,fx(x),x,deriv(x))
plt.plot(localmin,deriv(localmin),'ro')
plt.plot(localmin,fx(localmin),'ro')
plt.xlim(x[[0,-1]])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.legend(['y','dy'])
plt.show()
```

```python

localmin = np.random.choice(x,1)

learning_rate = .001
training_epoches =1000

modelparams = np.zeros((training_epoches,2))
for i in range(training_epoches):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad
    # print(localmin,grad)
    # print(modelparams[i])
    modelparams[i] = localmin[0],grad[0]

# print(modelparams)
print(localmin)
```

### x值随迭代的变化图

```python
### 梯度和迭代之间的关系图
fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    ax[i].plot(modelparams[:,i],'o-')    
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'Final estimated minimum {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')
plt.show()

```
