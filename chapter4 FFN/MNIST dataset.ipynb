```python


import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

```python

data = np.loadtxt(open("../data/mnist_train_small.csv",'r'),delimiter=',')
print(data.shape)
```

```python

labels = data[:,0]
data = data[:,1:]

print(labels.shape)
print(labels.shape)
```

```python
labels
```

```python

fig,axs = plt.subplots(3,4,figsize=(10,6))

for ax in axs.flatten():
    
    randing2show = np.random.randint(0,high= data.shape[0])
    
    img = np.reshape(data[randing2show,:],(28,28))
    ax.imshow(img,cmap='gray')
    ax.set_title('The number %i'%labels[randing2show])


plt.suptitle('How humans see the data',fontsize=20)

plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```
