

 深度解析协方差
 
 协方差是一个统计量，用于衡量两个随机变量之间的关系。它的定义是：
 
 协方差(X, Y) = E[(X - E[X])(Y - E[Y])]

 
 其中，E表示期望值。协方差的值可以是正的、负的或零：
 - 正协方差：表示两个变量之间存在正相关关系，即一个变量增加时，另一个变量也倾向于增加。
 - 负协方差：表示两个变量之间存在负相关关系，即一个变量增加时，另一个变量倾向于减少。
 - 零协方差：表示两个变量之间没有线性关系。
 
 协方差的作用：
 1. 用于描述数据集中的变量之间的关系。
 2. 在主成分分析（PCA）中，协方差矩阵用于确定数据的主要成分。
 3. 在金融领域，协方差用于评估资产之间的风险和收益关系。
 
 举例说明：
 假设我们有两个变量：身高（X）和体重（Y）。如果我们发现身高和体重之间的协方差为正值，说明身高较高的人通常体重也较重；如果协方差为负值，说明身高较高的人通常体重较轻；如果协方差为零，说明身高和体重之间没有明显的线性关。


```python

import numpy as np
X  = [ 2,4,6,8,10]
Y = [3,7,5,10,9]


Z = np.array([X,Y]).T
print(Z)
print(Z.shape)


covariance_matrix = np.cov(Z, rowvar=False)
print("Covariance Matrix:\n", covariance_matrix)



```

```python

HW= np.array([
    [90,32],
    [110,44],
    [170,65],
    [180,75],])

# print(HW.T)

COV_HW = np.cov(HW.T)

print(COV_HW[0])
print(COV_HW[1])
cv = np.cov(HW[0],HW[1])
print(cv)
```

```python



import matplotlib.pyplot as plt

# 创建散点图
plt.figure(figsize=(8, 6))
plt.scatter(HW[:, 0], HW[:, 1], color='blue', label='Height-Weight Points')

# 添加箭头表示向量
for i in range(len(HW)):
    plt.arrow(0, 0, HW[i,0], HW[i,1], 
             head_width=2, head_length=2, 
             fc='red', ec='red', 
             alpha=0.5)

# 设置坐标轴标签
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# 添加标题和图例
plt.title('Height-Weight Vectors')
plt.legend()

# 添加网格
plt.grid(True, linestyle='--', alpha=0.7)

# 设置坐标轴范围，确保从原点开始
plt.xlim(0, max(HW[:,0])*1.1)
plt.ylim(0, max(HW[:,1])*1.1)

plt.show()

```
