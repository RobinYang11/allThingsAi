```python

# 创建示例数据
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])

# 使用 np.corrcoef 计算相关系数矩阵
correlation_matrix = np.corrcoef(x, y)

print(correlation_matrix)
```

### np.hstack 和 np.vstack
