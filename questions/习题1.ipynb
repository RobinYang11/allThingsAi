

### 判断题
 - 1. 反向传播本质上就是梯度下降 .
 - 2. 交叉熵一般用于输出概率的模型.
 - 3. 深度学习的整个目标就是找到最小化成本函数的权重集.
 - 4. 激活函数一般用于输入层，来激活输入.
 - 5. sigmoid函数的中心点一般为1. 大于1的为一类，小于1的为一类。
 - 6. 在ANN中，偏置的作用是为了更好地分类。
 - 7. 在ANN中，层越多，模型越精确。
 - 8. 在设计ANN之前，模型有多少层时明确的。
 - 9. 在ANN各个线性层之间增加一层非线性激活函数.
 - 10. 在训练模型时，学习率一般时固定的，且值为0.1.
 - 11. 维度越高，局部最小化情况越少.
 - 12. 在ANN 中，每一组权重值，都代表一个分离平面或者线，也叫决策边界。
