```python



import numpy as np
import matplotlib.pyplot as plt

# 设置时间范围
t = np.linspace(0, 5, 100)  # 0到5秒，100个点

# 计算下降距离 S = (1/2)gt^2，其中g=9.8m/s^2
g = 9.8
S = 0.5 * g * t**2

# 创建图像
plt.figure(figsize=(10, 6))
plt.plot(t, S, 'b-', linewidth=2)
plt.grid(True)

# 添加标签和标题
plt.xlabel('时间 t (s)')
plt.ylabel('下降距离 S (m)')
plt.title('自由落体运动的距离-时间关系图')

plt.show()
```
